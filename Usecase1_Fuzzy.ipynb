```python
import pandas as pd
from fuzzywuzzy import fuzz 
from fuzzywuzzy import process
```

```python
df1 = pd.read_csv(r"C:\Users\chiranjeevi.h\Downloads\free-7-million-company-dataset\companies_sorted.csv")
df2 = pd.read_csv(r"C:/Users/chiranjeevi.h/Downloads/naics-soc/naics_codes.csv" , encoding='unicode_escape')
```

```python
# Getting Accurate Naics code from the Records
d= []
def getNaicCode(df1, df2):
    read_count=0
    for index,companyDescription in df1.iterrows():
        for index,naicTitleCodes in df2.iterrows():
            accuracy = fuzz.token_set_ratio(companyDescription["industry"], naicTitleCodes["naics_title"])
            if read_count == 500:
                break
            if accuracy>90:
                read_count+=1
                d.append({'Company Name':companyDescription["name"] , 'NaicsCode':naicTitleCodes["naics_code"]})
                print(companyDescription["industry"],"-",naicTitleCodes["naics_code"],"-",companyDescription["name"])
                break
            #else:
              #  print("Nothing found")
        if read_count == 500:
            break
                
getNaicCode(df1,df2)
final_dataframe = pd.DataFrame(d ,columns=['Company Name' , 'NaicsCode'])
final_dataframe.to_csv(r"C:\Users\chiranjeevi.h\Downloads\free-7-million-company-dataset\companies_NaicsCodes10.csv", index=False)
```
